# Classifica칞칚o com o Dataset Tic-Tac-Toe Endgame 游꿢
Neste notebook, vamos prever se um jogo da velha (tic-tac-toe) termina com vit칩ria do jogador X, usando o estado final do tabuleiro como entrada.

**Modelos comparados:**
- 츼rvore de Decis칚o
- Regress칚o Log칤stica
- KNN

**Etapas:**
- Carga e codifica칞칚o do dataset
- Divis칚o treino/teste e normaliza칞칚o
- Treinamento e avalia칞칚o com F1-score e Acur치cia

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score
```

```python
# Carregar dataset direto da UCI (via GitHub mirror)
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/tic-tac-toe/tic-tac-toe.data'
columns = ['top-left', 'top-middle', 'top-right', 'middle-left', 'middle-middle', 'middle-right', 'bottom-left', 'bottom-middle', 'bottom-right', 'target']
df = pd.read_csv(url, header=None, names=columns)
df.head()
```

```python
# Transformar o alvo para bin치rio: positive = 1, negative = 0
df['target'] = df['target'].map({'positive': 1, 'negative': 0})
# Codificar os valores simb칩licos (x, o, b)
X = pd.get_dummies(df.drop(columns='target'))
y = df['target']
# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
# Normalizar
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Modelos
dt = DecisionTreeClassifier(random_state=42)
knn = KNeighborsClassifier()
lr = LogisticRegression(max_iter=1000)
# Treinar
dt.fit(X_train_scaled, y_train)
knn.fit(X_train_scaled, y_train)
lr.fit(X_train_scaled, y_train)
# Prever
y_pred_dt = dt.predict(X_test_scaled)
y_pred_knn = knn.predict(X_test_scaled)
y_pred_lr = lr.predict(X_test_scaled)
# Avalia칞칚o
results = pd.DataFrame({
    'Modelo': ['츼rvore de Decis칚o', 'KNN', 'Regress칚o Log칤stica'],
    'Acur치cia': [
        accuracy_score(y_test, y_pred_dt),
        accuracy_score(y_test, y_pred_knn),
        accuracy_score(y_test, y_pred_lr)
    ],
    'F1-Score': [
        f1_score(y_test, y_pred_dt),
        f1_score(y_test, y_pred_knn),
        f1_score(y_test, y_pred_lr)
    ]
})
results
```
